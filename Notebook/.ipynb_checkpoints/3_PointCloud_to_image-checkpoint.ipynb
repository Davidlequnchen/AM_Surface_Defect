```python
# ---- using numpy to print all the points -------
points_np_array = np.asarray(pcd.points)
print("the points converted to numpy are: \n")
print(points_np_array)
#------extract x,y,z value of the points-----
# # ------------standarization-----------
# bbdiag = float(np.linalg.norm(points_np_array.max(0) - points_np_array.min(0), 2))
# points_np_array = (points_np_array - points_np_array.mean(0))/bbdiag
# print("the points are: \n")
# print(points_np_array)
```

```python
scale = 1
x = points_np_array[:,0]* scale ## from m to mm
y = points_np_array[:,1]*scale  ## from m to mm
points_np_array[:,2] = points_np_array[:,2] - np.amin(points_np_array[:,2]) 
z = points_np_array[:,2]*scale  ## from m to mm
print("the z value extracted:" )
print (z)

# --------mean height-------------------
z_mean = np.mean(z)
print("Z mean: " + str(z_mean))

z_min = np.amin(z)
print ("Z minimum: "+ str(z_min))
print ("Z max: "+ str(np.amax(z)))
```

```python
x.shape[0]
```

```python
# tune the segmentation line
# z_mean = z_mean + 0.17

def z_function(x, y):
    # ax+by+cz = d ---> z = -(ax+by-d)/c
    return  z_mean + 0 * x + 0 * y       

# create a plane with meshgrid (mainly for visualization purpose)
x_plane = np.linspace(-11, 11, 30)
y_plane = np.linspace(-21, 21, 30)
# x_plane = np.linspace(-40, 40, 30)
# y_plane = np.linspace(-40, 40, 30)

X, Y = np.meshgrid(x_plane, y_plane)
Z = z_function(X, Y)
```

### Convert XYZ point cloud to grayscale image
Reference: https://stackoverflow.com/questions/53700089/convert-xyz-point-cloud-to-grayscale-image

```python
x_range=((df.X.max()-df.X.min()))
y_range=((df.Y.max()-df.Y.min()))
z_range=((df.Z.max()-df.Z.min()))
grid_x, grid_y = np.mgrid[df.X.min():df.X.max():(x_range*100j), df.Y.min():df.Y.max():(y_range*100j)]
points = df[['X','Y']].values
values = df['Z'].values / z_range * 255
# values = df['Z'].values * 100

grid_z0 = scipy.interpolate.griddata(points, values, (grid_x, grid_y), method='linear').astype(np.uint8)

fig, ax = plt.subplots(figsize=(14, 7))
# cmap = mpl.cm.jet  ## other options: cool, jet, gray, winter
p = ax.imshow(grid_z0, cmap='jet')

#legend
cbar = plt.colorbar(p, fraction=0.025, pad=0.05)
cbar.set_label('Dimension Deviation', rotation=270, labelpad=22, fontsize = 18)

for t in ax.xaxis.get_major_ticks(): t.label.set_fontsize(14)
for t in ax.yaxis.get_major_ticks(): t.label.set_fontsize(14)
# cbar.ax.set_yticklabels(['< -1', '0', '> 1'])
cbar.ax.tick_params(labelsize=12, pad = 8) 
```

### Isothermal lines for dimension deviation levels

```python
THRESHES = [120, 184, 220]
for val in THRESHES:
    ret, thresh = cv2.threshold(grid_z0, val, 255, 0)
    contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    cv2.drawContours(image=grid_z0, contours=contours, contourIdx=-1, color=(255, 255, 255), 
                     thickness=5, lineType=cv2.LINE_AA)

    
plt.figure(figsize=(14, 8))
p = plt.imshow(grid_z0, cmap='jet')
plt.colorbar(p)
# plt.savefig("plot_result/colour_isothermal_line_1")
```

```python
from sklearn.feature_extraction import image
from sklearn.cluster import spectral_clustering

mask = grid_z0.astype(bool)
img = grid_z0.astype(float)
print ("----------------------")

graph = image.img_to_graph(img, mask=mask)
graph.data = np.exp(-graph.data / graph.data.std())
print ("----------------------")

labels = spectral_clustering(graph, n_clusters=4, eigen_solver="arpack")
label_im = np.full(mask.shape, -1.0)
label_im[mask] = labels

plt.matshow(img)
plt.matshow(label_im)

plt.show()
```
